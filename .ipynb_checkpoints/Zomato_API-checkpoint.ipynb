## Details of zomato.csv -


**Restaurant Id**: Unique id of every restaurant across various cities of the world
**Restaurant Name** : Name of the restaurant
**Country Code** : Country in which restaurant is located
**City** : City in which restaurant is located
**Address** : Address of the restaurant
**Locality** : Location in the city
**Locality Verbose** : Detailed description of the locality
**Longitude** : Longitude coordinate of the restaurant's location
**Latitude** : Latitude coordinate of the restaurant's location
**Cuisines** : Cuisines offered by the restaurant
**Average Cost for two** : Cost for two people in different currencies
**Currency** : Currency of the country
**Has Table booking** : yes/no
**Has Online delivery** : yes/ no
**Is delivering** : yes/ no
**Switch to order menu** : yes/no
**Price range** : range of price of food
**Aggregate Rating** : Average rating out of 5
**Rating color**: depending upon the average rating color
**Rating text**: text on the basis of rating of rating
**Votes**: Number of ratings casted by people

## 1.The dataset is highly skewed toward the cities included in Delhi-NCR. So, we will summarise all the other cities in Rest of India while those in New Delhi, Ghaziabad, Noida, Gurgaon, Faridabad to Delhi-NCR. Doing this would make our analysis turn toward Delhi-NCR v Rest of India.

### 1.1 Plot the bar graph of number of restaurants present in Delhi NCR vs Rest of India.


```python
#1.1
#copying the data 
df=data.copy()
#grouping delhi ncr together
delhi_ncr=df[(df["Country Code"]==1)&((df["City"]=="New Delhi")|(df["City"]=="Ghaziabad")|(df["City"]=="Noida")|(df["City"]=="Gurgaon")|(df["City"]=="Faridabad"))]
#grouping non delhi ncr together
non_delhi=df[(df["Country Code"]==1)&((df["City"]!="New Delhi")&(df["City"]!="Ghaziabad")&(df["City"]!="Noida")&(df["City"]!="Gurgaon")&(df["City"]!="Faridabad"))]
import matplotlib.pyplot as plt
#Drawing the bar graph using pyplot
plt.bar(["Delhi-NCR","Non Delhi"],[len(delhi_ncr),len(non_delhi)])
plt.title("Delhi-ncr vs Non-Delhi ncr")
plt.show()
print("Number of restaurants in delhi :-",len(delhi_ncr))
print("Number of restaurants in non-delhi:-",len(non_delhi))
# print(len(delhi_ncr),len(non_delhi))
```

### 1.2 Find the cuisines which are not present in restaurant of Delhi NCR but present in rest of India.Check using Zomato API whether this cuisines are actually not served in restaurants of Delhi-NCR or just it due to incomplete dataset.


```python
#1.2

delhi_cuisines=set()
non_delhi_cuisines=set()
# function to split the cuisines in delhi ncr and collecting each cuisines
def delhi_cus(ele):
    ele=ele.split(",")
    for i in ele:
        i=i.strip()
        delhi_cuisines.add(i)
#function to split the cuisines in non delhi ncr and collecting each cuisines
def non_delhi_cus(ele):
    ele=ele.split(",")
    for i in ele:
        i=i.strip()
        non_delhi_cuisines.add(i)
non_delhi["Cuisines"].apply(non_delhi_cus)
delhi_ncr["Cuisines"].apply(delhi_cus)
#1.2
#printing the cuisines that are present in restof india and not present in delhi ncr
print("Cuisines that are present in rest of India and not in Delhi NCR")
print("---------------------------------------------------------")
for i in non_delhi_cuisines:
    if i not in delhi_cuisines:
        print(i)
        
        
        
#*********************************************************************************
#****************************conclusion******************************************
#*********************************************************************************        

#conclusion:--using zomato api the cuisines like malwani, canjun , bbq and german are present
#in delhi . The error is because of incomplete dataset
```

### 1.3 Find the top 10 cuisines served by maximum number of restaurants in Delhi NCR and rest of India.


```python
#1.3
#1.3
#copying data into df
df=data.copy()
delhi_cuisines_count={}
non_delhi_cuisines_count={}

#function to split cuisines in delhi and collecting them
def delhi_cus_count(ele):
    ele=ele.split(",")
    for i in ele:
        i=i.strip()
        delhi_cuisines_count[i]=delhi_cuisines_count.get(i,0)+1


#function to split cuisines in delhi and collecting them        
def non_delhi_cus_count(ele):
    ele=ele.split(",")
    for i in ele:
        i=i.strip()
        non_delhi_cuisines_count[i]=non_delhi_cuisines_count.get(i,0)+1
non_delhi["Cuisines"].apply(non_delhi_cus_count)
delhi_ncr["Cuisines"].apply(delhi_cus_count)


from collections import Counter


#collecting top 10 cuisines in delhi ncr
top_10_delhi_cus=Counter.most_common(delhi_cuisines_count)

#collection top 10 cuisines in non -delhi ncr
top_10_non_delhi_cus=Counter.most_common(non_delhi_cuisines_count)
top_10_delhi_cus[:10],top_10_non_delhi_cus[:10]



print("Top 10 cuisines in Delhi are :")
print("----------------------------------------------")
plot_x_delhi=[]
plot_y_delhi=[]
for i in top_10_delhi_cus[:10]:
    print(i[0],i[1])
    plot_x_delhi.append(i[0])
    plot_y_delhi.append(i[1])
print()
print()
plot_x_non_delhi=[]
plot_y_non_delhi=[]
print("Top 10 cuisines in NON-Delhi are :")
print("----------------------------------------------")
for i in top_10_non_delhi_cus[:10]:
    print(i[0],i[1])
    plot_x_non_delhi.append(i[0])
    plot_y_non_delhi.append(i[1])
```

## 2.2 Find the weighted restaurant rating of each locality and find out the top 10 localities with more weighted restaurant rating?


### 2.2.1 Weighted Restaurant Rating= (number of votes * rating) /  (number of votes) .


```python
# 2.2.1
#2nd answer 2.1
df=data.copy()

#***********************************************************
#***********************************************************
#I have done this on total dataset (Not in only india but in whole world)
#weighted aggreagate rating finding 
df["vot_rat"]=df["Votes"]*df["Aggregate rating"]
df_locality=df.groupby("Locality").sum()["vot_rat"]/df.groupby("Locality").sum()["Votes"]

#collecting top 
print("Top 10 locality with highest rating :---")
print("---------------------------------------------------")
df_locality_top=Counter.most_common(df_locality)
for i in df_locality_top[:10]:
    print(i)
```

### 3.4 Plot the pie graph of top 10 cuisines present in restaurants in the USA.

```python
# 3.4
#copying original data into df
df=data.copy()


#collecting USA country data
df=df[df["Country Code"]==216]
df["Cuisines"].dropna(inplace=True)
cuisines_usa={}


#function to split cuisines seperated by 'c' comma
def cuisines(ele):
    ele=ele.split(",")
    for i in ele:
        i=i.strip()
        cuisines_usa[i]=cuisines_usa.get(i,0)+1
df["Cuisines"].apply(cuisines)


from collections import Counter


#code to collect the top 10 restaurants
top_cuisines_usa=Counter.most_common(cuisines_usa)
print("Top 10 cuisines in USA are :")
x_food_name=[]
y_no_of_times=[]
for i in top_cuisines_usa[:10]:
    print(i[0],i[1])
    x_food_name.append(i[0])
    y_no_of_times.append(i[1])
# df["Cuisines"]

#code to plot the piechart 
plt.pie(y_no_of_times,labels=x_food_name,autopct="%.1f%%")
plt.axis("equal")
plt.title("Top 10 cuisines in USA")
plt.show()
```
